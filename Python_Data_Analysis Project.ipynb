```python
import matplotlib as plt
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['brand_name'], df['revenue'],color='skyblue')
plt.title('Total Revenue by Brand')
plt.xlabel('Brand Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

##### what is the total revenue for each Category Name ?

```python
df.groupby(['category_name'])['revenue'].sum()
```

```python
import matplotlib.pyplot as plt

df_sorted = df.sort_values(by='revenue', ascending=False)

plt.figure(figsize=(10, 6))
bars = plt.bar(df_sorted['category_name'], df_sorted['revenue'], color='skyblue')

plt.title('Total Revenue by Brand')
plt.xlabel('Brand Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
```
